```python
import os

# groundtruth 文件夹路径，替换成你自己的路径
ROOT_DIR = "/mnt/hdd2/Projects_dai/Projects_dai/MS-TCN2/data/surgery_I3D_new/groundtruth"

def replace_spaces_in_labels(root_dir: str):
    for fn in sorted(os.listdir(root_dir)):
        if not fn.endswith(".txt"):
            continue
        path = os.path.join(root_dir, fn)
        # 读入所有行
        with open(path, "r", encoding="utf-8") as f:
            lines = f.read().splitlines()

        # 替换每行中的空格为 _
        new_lines = [line.replace(" ", "_") for line in lines]

        # 写回文件（原地覆盖）
        with open(path, "w", encoding="utf-8") as f:
            f.write("\n".join(new_lines))
        print(f"Processed {fn}: {len(lines)} lines → {len(new_lines)} lines")

if __name__ == "__main__":
    replace_spaces_in_labels(ROOT_DIR)
    print("✅ All done.")
```

```python
for i in range(1,6):
    input_path = f"/mnt/hdd2/Projects_dai/Projects_dai/MS-TCN2/data/surgery_I3D_new/splits/train.split{i}.bundle"
    output_path = f"/mnt/hdd2/Projects_dai/Projects_dai/MS-TCN2/data/surgery_I3D_new/splits/train_argumented.split{i}.bundle"
    with open(input_path,'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    argumented_lines = []
    for line in lines:
        argumented_lines.append(line)
        argumented_lines.append(f'{line}_argumented')
    with open(output_path,'w') as f:
        for line in argumented_lines:
            f.write(line + '\n')
    print(f'finished')
```
